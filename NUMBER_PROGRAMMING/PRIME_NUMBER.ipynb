```python
#METHOD-1

num=23
count=0
if num>1:
    for val in range(1,num+1):
        if num % val==0:
            count+=1
    if count==2:
        print("Prime Number")
    else:
        print("Not Prime Number")
else:
    print("Not prime Number")
```

```python
#METHOD-2

num=124
if num>1:
    for val in range(2,num//2+1):
        if num % val==0:
            print("Not Prime Number")
            break
    else:
        print("Prime Number")
else:
    print("Not prime Number")
```
