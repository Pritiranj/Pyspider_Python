```python
#REVERSE_NUMBER

#METHOD-1

num=173
rev=0

while num>0: 
    rem=num%10
    rev=rem + rev*10
    num//=10
print(rev)
```

```python
num=173
rev=0
power=10**(len(str(num))-1)
while num>0:
    rem=num%10
    rev=rem + rev*10
    power//=10
    num//=10
print(rev)
```

```python
num=122
dup=num
rev=0

while num>0:
    rem=num%10
    rev=rem+rev*10
    num//=10
if dup==rev:
    print('Palindro_Number')
else:
    print('Not Palindrome_number')
```
