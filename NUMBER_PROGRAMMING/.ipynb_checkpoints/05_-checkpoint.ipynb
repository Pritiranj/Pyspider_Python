```python
#DISARUG NUMBER
#EX-135=1^1+3^2+5^3=135

num=13
dup=num
temp=num
res=0
count=0
while num!=0:
    num//=10
    count+=1
while dup !=0:
    rem=dup%10
    res=res+rem**count
    count-=1
    dup//=10
if temp==res:
    print("Disarug no")
else:
    print("not")
```

```python
#SPECIAL NUMBER
#EX= 28=1,2,4,7,14,28
# SUM OF FACTORAL EXCEPT THAT NUMBER WILL BE THAT NUMBER
#1+2+4+7+14=28//


num=30
res=0
for val in range(1,num//2+1):
    if num%val==0:
        res+=val
if res==num:
    print("Special_number")
else:
    print("Not a Special Number")
    
```

```python
#STRONG NUMBER
#1!+4!+5!=1+24+120=145
#SUM OF FACTORIAL OF DIGIT OF A NUMBER WILL BE SAME

num=145
dup=num
res=0
fact=1
while num!=0:
    rem=num%10
    fact=1# reset the fact value for each digit
    for val in range (1,rem+1):
        fact*=val
    res=res+fact
    num//=10
    
if dup==res:
    print("strong")
else:
    print("not strong")
```
