```python
#FACTORLA NUMBER

#METHOD-1

num=5
fact_num=1

for val in range(1,num+1):
    fact_num*=val
print(fact_num)
```

```python
#METHOD-2
num=5
val,fact=1,1
while val!=num+1:
    fact*=val
    val+=1
print(fact)
    
```

```python
#COUNT NUMBER OF DIGIT 

num=1234
count=0
while num>0:
    num//=10
    count+=1
print(count)
        
```
