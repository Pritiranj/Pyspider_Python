```python
def fact(num,facta=1):
    while num != 0:
        facta*=num
        num-=1
    return facta

def adding(num,res=0):

        while num != 0:
            rem = num % 10
            res=res+fact(rem)
            num//=10
        return res
def strong (num):
    if adding(num) == num :
        return True
    return False
num=145
print(strong(num))
```
