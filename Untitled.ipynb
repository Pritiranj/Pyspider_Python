```python
def facto(num,res=1):
    yield 1
    for val in range(1,num+1):
        res*=val
        yield res
genobj=facto(5)
print(next(genobj))
print(next(genobj))
print(next(genobj))
print(next(genobj))
print(next(genobj))
```
