```python
def outer(arg):
    l=[]
    def inner():
        if len(l)==0:
            obj=arg()
            l.append(obj)
        return l[0]
    return inner
@outer
class sun:
    def __init__(self):
        print('hello')
si=sun()
print(si)
print(si)
```
