```python
num=19
dup=num
rev=0

while num !=0:
    rem=num%10
    rev=rev*10+rem
    num//=10
if dup != rev:
    if dup >1:
        for val in range(2,dup//2+1):
            if dup % val ==0:
                print('nen')
                break
        else:
            if rev>1:
                for val in range(2,rev//2+1):
                    if rev%val==0:
                        print('na')
                        break
                else:
                    print('emp')
    else:
        print('nem')
else:
    print('nem')
```

```python
def half(l):
    if len(l)>1:
        mid=len(l)//2
        left=l[:mid]
        right=l[mid:]
        half(left)
        half(right)
        merge(l,left,right)
    return l

def merge(l,left,right):
    lind,mind,rind=0,0,0
    while lind< len(left) and rind<len(right):
        if left[lind] <right[rind]:
            l[mind]=left[lind]
            lind+=1
        else:
            l[mind]=right[rind]
            rind+=1
        mind+=1
    while lind <len(left):
        l[mind]=left[lind]
        lind+=1
        mind+=1
    while rind<len(right):
        l[mind]=right[rind]
        mind+=1
        rind+=1
l=[1,2,3,9,8,7,0,-1]
print(half(l))
```

```python
def con(num,power=1):
    if num==0:
        return 0
    return (num%2)*power+con(num//2,power*10)
num=5
print(con(num))
    
```

```python
def con(num,power=1):
    if num==0:
        return 0
    return (num%10)*power+con(num//10,power*2)
num=101
print(con(num))
    
```

```python
def con(num,power=1,res=0):
    while num != 0:
        rem=num%2
        res=rem*power+res
        power*=10
        num//=2
    return res
num=5
print(con(num))
```

```python
def inter(l,user,low=0 ,high=len(l)-1):
    # if high is None:
    #     high=len(l)-1
    while low <= high and l[low]<= user <= l[high]:
        mid=int(low+((high-low)/(l[high]-l[low]))*(user-l[low]))
        if l[mid]<user:
            low=mid+1
        elif l[mid]>user:
            high=mid-1
        elif l[mid]==user:
            return mid
    return -1
l=[1,2,0,90,39,67,56]
user=90
print(inter(l,user))
```

```python
def outer(arg):
    l = []  # Holds the instance
    
    def inner():
        if len(l) == 0:  # Create instance only once
            obj = arg()  # Call class constructor to create instance
            l.append(obj)  # Store instance in the list
        return l[0]  # Return the same instance every time
    
    return inner

@outer
class Sun:
    def __init__(self):
        print('hello')
    
    def m1(self):
        print('hi')

# Calling the class and checking if only one instance is created
obj = Sun()  # First call, prints 'hello'
obj2 = Sun()  # Will not print 'hello', returns same instance as obj
obj3 = Sun()  # Will not print 'hello', returns same instance as obj
print(obj is obj2)  # Should print True, as all refer to the same instance
print(obj is obj3)  # Should print True, as all refer to the same instance
```
