```python
#singleton class

def outer(arg):
    res=[]
    def inner():
        if len(res)==0:
            obj=arg()
            res.append(obj)
        return res[0]
    return inner
@outer
class sun:
    def __init__(self):
        print('hello')
ob1=sun()
ob2=sun()
print(ob1)
print(ob2)
```

```python
#decorator

def outer(arg):
    def inner():
        obj=arg()
        return obj
    return inner
@outer
class sun:
    def __init__(self):
        print('hello')
ob1=sun()
ob2=sun()
print(ob1)
print(ob2)
print(ob1)
```

```python
#overloading

class sun:
    def m1(self,n1):
        print(n1)
    def m1(self,n1,n2):
        print(n1*n2)
    def m1(self,n1,n2,n3):
        print(n1*n2*n3)
obj=sun()
obj.m1(1,2)
```

```python
#overridding

class sun:
    def mock(self):
        print('yes')
class sat:
    def mock(self):
        print('no')
obj=sat()
obj.mock()
```
