```python
def fact(num):
    if num<2:
        return 1
    return num*fact(num-1)
def add(num):
    if num==0:
        return 0
    return fact(num%10)+add(num//10)
def strong(num):
    if num ==add(num):
        return True
    return False
num=153
print(list(filter(strong,range(1,200))))
count=0
res=[]
for ele in range(1,201):
    if strong(ele):
        if count %2 ==0:
            res+=[ele]
        count+=1
print(res)
```

```python
def prime(num):
    if num >1:
        for val in range(2,num//2+1):
            if num % val ==0:
                return False
        return True
    return False
def palin(num,dup,res=0):
    while num != 0 :
        rem= num %10
        res=rem+res*10
        num//=10
    return res==dup
def poly(num):
    if prime(num) and palin(num,num):
        return True
    return False
print(list(filter(poly,range(1,1000))))
res=[]
count=0
for ele in range(1,1000):
    if poly(ele):
        if count % 2==0:
            res+=[ele]
        count+=1
        
print(res)
        
```

```python
l=[1,28,9,8,78,67]
user=78
l=sorted(set(l))
print(l)
def inter(l,user,low=0,high=len(l)-1):
    while low <= high and l[low] <= user <= l[high]:
        
        mid= int(low+(high-low)//(l[high]-l[low])*(user-l[low]))
        if user<l[mid]:
             high=mid-1
        elif user > l[mid]:
             low=mid+1
        elif user==l[mid]:
             return mid
    else:
        return -1
print(inter(l,user))            
```

```python
#bubble sort

l=[1,2,9,8,0,7,6,-8,0]
for ind1 in range(len(l)-1,0,-1):
    for ind2 in range(ind1):
        if l[ind2]>l[ind2+1]:
            l[ind2],l[ind2+1]=l[ind2+1],l[ind2]
print(l)
```

```python
#selection
l=[1,2,9,8,0,7,6,-8,0]
for ind1 in range(len(l)-1):
    li=ind1
    for ind2 in range(ind1+1,len(l)):
        if l[li]>l[ind2]:
            li=ind2
    l[li],l[ind1]=l[ind1],l[li]
print(l)

```

```python
#insertion sort
l=[1,2,9,8,0,7,6,-8,0]
for ind in range(1,len(l)):
    h=ind
    val=l[h]
    while h >0 and val < l[h-1]:
        l[h]=l[h-1]
        h-=1
    l[h]=val
print(l)
```

```python
#quick sort
def quick(l):
    if len(l)<=1:
        return l
    pivot=l[0]
    left=[ele for ele in (l[1:]) if ele <pivot]
    middle = [ele for ele in l if ele == pivot]
    right=[ele for ele in (l[1:]) if ele >pivot]
    return quick(left)+middle+quick(right)
l=[1,2,9,8,0,7,6,-8,0]
print(quick(l))

```

```python
#decorator
def outer(arg):
    def inner():
        obj=arg()
        return obj
    return inner
@outer
class v:
    def __init__(self):
        print('hello')
a=v()
a1=v()
```

```python
#singleton
def outer(arg):
    l=[]
    def inner():
        if len(l)==0:
            obj=arg()
            l.append(obj)
        return l[0]
    return inner
@outer
class v:
    def __init__(self):
        print('hi')
a=v()
b=v()
v=v()
```

```python
#merge sort
def half(l):
    if len(l)>1:
        mid=len(l)//2
        left=l[:mid]
        right=l[mid:]
        half(left)
        half(right)
        merge(l,left,right)
    return l
def merge(l,left,right):
    lind,rind,mind=0,0,0
    while lind < len(left) and rind <len(right):
        if left[lind] <right[rind]:
            l[mind]=left[lind]
            lind+=1
        else:
            l[mind]=right[rind]
            rind+=1
        mind+=1
    while lind <len(left):
        l[mind]=left[lind]
        lind+=1
        mind+=1
    while rind <len(right):
        l[mind]=right[rind]
        rind+=1
        mind+=1
l=[1,2,9,0,8,-1,-3]
half(l)
print(l)
```

```python
#overloading
class v1:
    def m1(self,n1):
        print(n1)
    def m1(self,n1,n2):
        print(n1*n2)
obj=v1()
obj.m1(2,2)
```

```python
#achive overloading
class v:
    def m1(self,*args):
        if len(args)>0:
            res=1
            for ele in args:
                res*=ele
            print(res)
        else:
            print(0)
obj=v()
obj.m1(2)
```

```python
#overriding
class sun:
    def m1(self):
        print('sunday')
class sat(sun):
    def m1(self):
        print('saturday')
obj=sat()
obj.m1()
```
