```python
#bubble sort

l=[9,10,0,-1,2,-3,8,-4,1]

for ind1 in range(len(l)-1,0,-1):
    for ind2 in range(ind1):
        if l[ind2] >l[ind2+1]:
            l[ind2],l[ind2+1]=l[ind2+1],l[ind2]
print(l)
```

```python
#selection sort
l=[9,10,0,-1,2,-3,8,-4,1]
for ind1 in range(len(l)-1):
    li=ind1
    for ind2 in range(ind1+1,len(l)):
        if l[li]>l[ind2]:
            li=ind2
    else:
        l[ind1],l[li]=l[li],l[ind1]
print(l)
```

```python
#quick sort

l=[9,10,0,-1,2,-3,8,-4,1]

def quick(l):
    if len(l)<=1:
        return l
    pivot=l[0]
    left=[ele for ele in (l[1:]) if pivot>ele]
    right=[ele for ele in (l[1:]) if ele>pivot]
    return quick(left)+[pivot]+quick(right)
print(quick(l))
```

```python
#meege sort

def half(l):
    if len(l)>1:
        mid=len(l)//2
        left=l[:mid]
        right=l[mid:]
        half(left)
        half(right)
        merge(l,left,right)
    return l
def merge(l,left,right):
    lind,rind,mind=0,0,0
    while lind < len(left) and rind <len(right):
        if left[lind]<right[rind]:
            l[mind]=left[lind]
            lind+=1
        else:
            l[mind]=right[rind]
            rind+=1
        mind+=1

    while lind<len(left):
        l[mind]=left[lind]
        lind+=1
        mind+=1
    while rind<len(right):
        l[mind]=right[rind]
        rind+=1
        mind+=1
l=[9,10,0,-1,2,-3,8,-4,1]
print(half(l))
```

```python
#insertion sort

l=[9,10,0,-1,2,-3,8,-4,1]
for ind in range(1,len(l)):
    h=ind
    val=l[h]
    while h>0 and val<l[h-1]:
        l[h]=l[h-1]
        h-=1
    l[h]=val
print(l)
```

```python
#constructor chaining

class v1:
    def __init__(self,name,mob):
        self.name=name
        self.mob=mob
    def details(self):
        print(self.name)
        print(self.mob)
class v2(v1):
    def __init__(self,name,mob,email,gender):
        super().__init__(name,mob)
        self.email=email
        self.gender=gender
    def details(self):
        v1.details(self)
        print(self.email)
        print(self.gender)
obj=v2('situ',7896,'bpritu@','male')
obj.details()
```
