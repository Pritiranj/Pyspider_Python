```python


def half(l):
    if len(l)>1:
        mid=len(l)//2
        left=l[:mid]
        right=l[mid:]
        half(left)
        half(right)
        merge(l,left,right)
def merge(l,left,right):
    mind,lind,rind=0,0,0

    while lind<len(left) and rind<len(right):
        if left[lind]<right[rind]:
            l[mind]=left[lind]
            lind+=1
        else:
            l[mind]=right[rind]
            rind+=1
        mind+=1
    while lind<len(left):
        l[mind]=left[lind]
        lind+=1
        mind+=1
    while rind<len(right):
        l[mind]=right[rind]
        rind+=1
        mind+=1
l=[1,2,5,7,4,0,3]
half(l)
print(l)
```

```python
l=[1,2,5,7,4,0,3]
def quick(l):
    if len(l)<=1:
        return l
    pivot=l[0]
    left=[ele for ele in (l[1:]) if pivot>ele]
    right=[ele for ele in (l[1:]) if ele>pivot]
    return quick(left)+[pivot]+quick(right)
print(quick(l))
```

```python
def prime(num,val=2):
    if num % val== 0:
        return False
    if val==num//2+1:
        return True
    return prime(num,val+1)

def rev(num,res=0):
    if num ==  0 :
        return res
    res=res*10+num%10
    return rev(num//10,res)
num=153
# print(rev(num))

def poly(num):
    sid=rev(num)
    if prime(num) and sid == num:
        return 'yes'
    return'no'
num=11
print(poly(num))
```
