```python
def outer(arg):
    res=[]
    def inner():
        if len(res)==0:
            obj=arg()
            res.append(obj)
        return res
    return inner

@outer
class sun:
    def __init__(self):
        print('hello')
ob=sun()
ob1=sun()
ob2=sun()
print(ob is ob1 is ob2)
```
