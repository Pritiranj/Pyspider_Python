```python
#BUBBLE SORT

l=[1,2,3,4,5,0,-8,9,-5]

for ind1 in range(len(l)-1,0,-1):
    for ind2 in range(ind1):
        if l[ind2]>l[ind2+1]:
            l[ind2],l[ind2+1]=l[ind2+1],l[ind2]
print(l)
```

```python
l=[1,2,3,4,5,0,-8,9,-5]
def quick(l):
    if len(l)<=1:
        return l
    pivot=l[0]
    left=[ele for ele in (l[1:]) if ele<pivot]
    right=[ele for ele in (l[1:]) if ele>pivot]
    return quick(left)+[pivot]+quick(right)
print(quick(l))
```
