```python
def palin(num):
    res=0
    dup=num
    while num != 0:
        rem=num%10
        res=rem+res*10
        num//=10
    return res==dup
print(list(filter(palin,range(1,100))))
print(list(ele for ele in range (1,100) if palin(ele)))
    
```

```python
def quick(l):
    if len(l)<1:
        return l
    pivot=l[0]
    left=[ele for ele in (l[1:]) if ele < pivot]
    right=[ele for ele in (l[1:]) if ele > pivot]
    return quick(left)+[pivot]+quick(right)


l = [1, 9, 8, 0, 6, 12, 76, -1, 0]
print(quick(l))
    
```
