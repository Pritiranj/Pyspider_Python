```python
#Armstrong Number

def arm(num,power):
    
    if num ==0:
        return 0
    return (num%10)**power+arm(num//10,power)
num=153
power=len(str(num))
if arm(num,power)==num:
    print('yes')
else:
    print('no')
```

```python
#bubble sort
s=[1,20,9,0,89,89,787,6778,28,19]

for ind1 in range(len(s)-1,0,-1):
    for ind2 in range(ind1):
        if s[ind2]>s[ind2+1]:
            s[ind2],s[ind2+1]=s[ind2+1],s[ind2]
print(s)
        
```

```python
def outer(arg):
    res=[]
    def inner():
        if len(res)==0:
            obj=arg()
            res.append(obj)
            
        return res[0]
    return inner

@outer
class sun():
    def __init__(self):
        print('hello')
    def m1(self):
        print('yes')
obj=sun()
obj.m1()
```
