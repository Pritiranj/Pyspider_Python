```python
#bubble sort

l=[1,2,3,4,-8,-6,19]

for ind1 in range(len(l)-1,0,-1):
    for ind2 in range(ind1):
        if l[ind2]>l[ind2+1]:
            l[ind2],l[ind2+1]=l[ind2+1],l[ind2]
print(l)
```

```python
#insertion sort

l=[1,8,9,0,-2,89,78]

for ind in range(1,len(l)):
    h=ind
    val=l[h]
    while h>0 and val<l[h-1]:
        l[h]=l[h-1]
        h-=1
    l[h]=val
print(l)
```

```python
#quick sort

def quick(l):
    if len(l)<1:
        return l
    pivot=l[0]
    left=[ele for ele in (l[1:]) if ele<pivot]
    right=[ele for ele in (l[1:]) if ele>pivot]
    return quick(left)+[pivot]+quick(right)

l=[1,2,9,0,8,-8,-7]
print(quick(l))
```

```python
#selection sort

l=[1,0,-9,0,7,6,9,89]

for ind1 in range(len(l)-1):
    li=ind1
    for ind2 in range(ind1+1,len(l)):
        if l[li]>l[ind2]:
            li=ind2
    l[ind1],l[li]=l[li],l[ind1]
print(l)
```
