```python
#singleton class

def outer(arg):
    res=[]
    def inner():
        if len(res)==0:
            obj=arg()
            res.append(obj)
        return res[0]
    return inner
@outer
class sun:
    def __init__(self):
        print('hello')
ob1=sun()
ob2=sun()
print(ob1)
print(ob2)
```

```python
#decorator

def outer(arg):
    def inner():
        obj=arg()
        return obj
    return inner
@outer
class sun:
    def __init__(self):
        print('hello')
ob1=sun()
ob2=sun()
print(ob1)
print(ob2)
print(ob1)
```

```python
#overloading

class sun:
    def m1(self,n1):
        print(n1)
    def m1(self,n1,n2):
        print(n1*n2)
    def m1(self,n1,n2,n3):
        print(n1*n2*n3)
obj=sun()
obj.m1(1,2)
```

```python
#constructor

class v1:
    def __init__(self,name,gen):
        self.name=name
        self.gen=gen
    def details(self):
        print(self.name)
        print(self.gen)
class v2(v1):
    def __init__(self,name,gen,email,mob):
        v1.__init__(self,name,gen)
        self.email=email
        self.mob=mob
    def details(self):
        v1.details(self)
        print(self.email)
        print(self.mob)

obj=v2('situ','male','email','7894544774')
obj.details()
```

```python
#user defind error

class arm(Exception):
    def __init__(self,spam):
        self.spam=spam
    def __str__(self):
        return self.spam
try:
    num=15
    dup=num
    res=0
    power=len(str(num))
    while num != 0:
        rem=num%10
        res=res+rem**power
        num//=10

    if dup != res:
        raise arm('not armstrong')
except Exception as msg:
    print(msg)
else:
    print('arm')
finally:
    print('executed')
```

```python
#iterator

class ten:
    def __init__(self):
        self.num=20
    def __iter__(self):
        return self
    def __next__(self):
        if self.num>36:
           raise StopIteration('no')
        else:
            val=self.num
            self.num+=5
            return val
    
obj=ten()
itobj=iter(obj)
print(itobj)
while True:
    print(next(itobj))
```
