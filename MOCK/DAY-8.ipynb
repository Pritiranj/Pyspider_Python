```python
def half(l):
    if len(l)>1:
        mid=len(l)//2
        left=l[:mid]
        right=l[mid:]
        half(left)
        half(right)
        merge(l,left,right)
    return l

def merge(l,left,right):
    lind,mind,rind=0,0,0
    while lind< len(left) and rind<len(right):
        if left[lind] <right[rind]:
            l[mind]=left[lind]
            lind+=1
        else:
            l[mind]=right[rind]
            rind+=1
        mind+=1
    while lind <len(left):
        l[mind]=left[lind]
        lind+=1
        mind+=1
    while rind<len(right):
        l[mind]=right[rind]
        mind+=1
        rind+=1
l=[1,2,3,9,8,7,0,-1]
print(half(l))
```

```python
def outer(arg):
    l = []  # Holds the instance
    
    def inner():
        if len(l) == 0:  # Create instance only once
            obj = arg()  # Call class constructor to create instance
            l.append(obj)  # Store instance in the list
        return l[0]  # Return the same instance every time
    
    return inner

@outer
class Sun:
    def __init__(self):
        print('hello')
    
    def m1(self):
        print('hi')

# Calling the class and checking if only one instance is created
obj = Sun()  # First call, prints 'hello'
obj2 = Sun()  # Will not print 'hello', returns same instance as obj
obj3 = Sun()  # Will not print 'hello', returns same instance as obj
print(obj is obj2)  # Should print True, as all refer to the same instance
print(obj is obj3)  # Should print True, as all refer to the same instance
```

```python
#selection sort
l=[1,2,9,8,6,0,-1]
for ind1 in range(len(l)-1):
    li=ind1
    for ind2 in range(ind1+1,len(l)):
        if l[ind2]<l[li]:
            li=ind2
    l[ind1],l[li]=l[li],l[ind1]
print(l)
```

```python
#quick sort
l=[1,2,9,8,6,0,-1]

def quick(l):
    if len(l)<1:
        return l
    pivot=l[0]
    left=[ele for ele in (l[0:]) if ele <pivot]
    right=[ele for ele in (l[0:]) if ele > pivot]
    return quick(left)+[pivot]+quick(right)
print(quick(l))
```
