```python
def gcd(num1,num2):
    if num1>num2:
        gcd=num2
    else:
        gcd=num1
    while True:
        if num1%gcd==0 and num2% gcd==0:
            return gcd
        gcd-=1
num1=24
num2=12
print(gcd(num1,num2))
```

```python
def gcd(num1,num2,val=None):
    if val==None:
        val=max(num1,num2)
    if val%num1==0 and val%num2==0:
        return val
    return gcd(num1,num2,val+1)
num1=24
num2=12
print(gcd(num1,num2))
```

```python
l=[1,29,8,78,0,7,8]
l=sorted(set(l))
user=78

print(l)

def inter(l,user,low=0,high=len(l)-1):
    while low <= high and l[low] <= user <= l[high]:
        mid=int(low+(high-low)/(l[high]-l[low])*(user-l[low]))
        if user<l[mid]:
            high=mid-1
        elif user>l[mid]:
            low=mid+1
        elif user==l[mid]:
            return mid
    else:
        return -1
print(inter(l,user))
```

```python
def half(l):
    if len(l)>1:
        mid=len(l)//2
        left=l[:mid]
        right=l[mid:]
        half(left)
        half(right)
        merge(l,left,right)
    return l
def merge(l,left,right):
    lind,mind,rind=0,0,0
    while lind < len(left) and rind < len(right):
        if left[lind] < right[rind]:
            l[mind]=left[lind]
            lind+=1
        else:
            l[mind]=right[rind]
            rind+=1
        mind+=1
    while lind < len(left):
        l[mind]=left[lind]
        lind+=1
        mind+=1
    while rind < len(right):
        l[mind]=right[rind]
        rind+=1
        mind+=1
l=[1,8,9,0,-2,3]
half(l)
print(l)
```
