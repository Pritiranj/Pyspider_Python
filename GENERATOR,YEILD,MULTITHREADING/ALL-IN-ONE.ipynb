```python
def fact(num, res=1):  # Changed default res to 1
    yield 1

    for ele in range(1, num + 1):
        res *= ele  # Multiply res by ele
        yield res   # Yield the factorial result

genobj = fact(5)

while True:
    try:
        print(next(genobj))
    except StopIteration:
        break
```

```python
import time
import threading

def sample1(n):
    for val in range(1, n):
        print('hello')
        time.sleep(1)

def sample2(m):
    for val in range(1, m):
        print('bye')

# Pass function references (no quotes)
t1 = threading.Thread(target=sample1, args=(5,), name='fun1')
t2 = threading.Thread(target=sample2, args=(5,), name='fun2')

# Start both threads
t1.start()
t2.start()

# Wait for both threads to complete before exiting
t1.join()
t2.join()

print("Both threads have finished.")

```
