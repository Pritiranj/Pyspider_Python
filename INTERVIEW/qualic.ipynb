```python
l=[1,9,8,0,-2]
for ind1 in range (len(l)-1,0,-1):
    for ind2 in range(ind1):
        if l[ind2]>l[ind2+1]:
            l[ind2],l[ind2+1]=l[ind2+1],l[ind2]
print(l)
```

```python
l=[1,9,0,8,-1,2,9]
for ind1 in range(len(l)-1):
    li=ind1
    for ind2 in range(ind1+1,len(l)):
        if l[ind2]<l[li]:
            li=ind2
    l[ind1],l[li]=l[li],l[ind1]
print(l)
```
