```python
#Q25

num = 5
for row in range(1, num + 1):
    
    for sp in range(num-row):
        print(" ", end=" ")
   
    for st in range(num - row + 1, num + 1):
        print(st, end=" ")
    print()  
```

```python
#26
num = 5
for row in range(num,0,-1):
    
    for sp in range(num-row):
        print(" ", end=" ")
   
    for st in range(num - row + 1, num + 1):
        print(st, end=" ")
    print()  
```
