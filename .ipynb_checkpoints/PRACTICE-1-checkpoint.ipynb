```python
def fact(num,power):
    if num==0:
        return 0
    return (num%10)**power+fact(num//10,power-1)
num=135
power=len(str(num))
print(fact(num,power))
    
    
```

```python
def fact(num):
    if num<=1:
        return 1
    return num* fact(num-1)
def add(num):
    if num==0:
        return 0
    return fact(num%10)+add(num//10)
def strong(num):
    if num==add(num):
        return 'yes'
    else:
        return 'no'
num=145
print(strong(num))
```

```python
def find(num):
    if num==0:
        return 0
    return (num%10)**2+find(num//10)
def check(num):
    if num>=9:
        return check((find(num)))
    return num
num=69
sid=check(num)
if sid==1 or sid==7:
    print('true')
else:
    print('no')
```

```python
def square(num,res=0):
    while num != 0:
        rem=num%10
        res=rem**2+res
        num//=10
    return res
def special(num):
    while num>9:
        num=square(num)
    return num
num=69
sid1=special(num)
if sid1==1 or sid1==7:
    print('true')
else:
    print('no')
```

```python
def fact(num,res=1):
    while num !=0:
        res*=num
        num-=1
    return res
for ele in map(fact,range(1,7)):
    print(ele)
```

```python
s='we have class on sunday'
l=[]
word=''
for ch in s:
    if ch ==' ':
        l=[word]+l
        word=''
    else:
        word=ch+word
l=[word]+l
ind=0
res=''
while ind !=len(l):
    if ind != len(l):
        res+=l[ind]+' '
    else:
        res+=l[ind]
    ind+=1
print(res)
        
        
```

```python
s='aaabbbcccdd'
count=1
res=''
for ind in range(len(s)-1):
    if s[ind]==s[ind+1]:
        count+=1
    else:
        res=res+str(count)+s[ind]
        count=1
else:
    res=res+str(count)+s[-1]
print(res)
        
```

```python
def find(l,target,ind=0):
    if ind ==len(l):
        return -1
    else:
        if l[ind]==target:
            return ind
    return find(l,target,ind+1)
l=[5,20,39,45]
target=3
print(find(l,target))
```
