```python
class CustomError(Exception):
    pass

class Sun:
    def __init__(self):
        self.num = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.num == 3:
            raise CustomError("Custom error at number 3")
        if self.num <= 5:
            val = self.num
            self.num += 1
            return val
        else:
            raise StopIteration
s = Sun()
for i in s:
    print(i)
```
