```python
with open('sample1.txt','w') as fobj:
    fobj.writelines('hello\nbye\nholliday\n')
    fobj.writelines(['13\n','4\n','5','6\n'])
    fobj.writelines(('abcd\n','mock'))
    fobj.writelines({'a:30','b:400'})
```

```python
with open('sample1.txt','w') as fobj:
    print(fobj.tell())
    print(fobj.readable(),fobj.writable())
    fobj.write('we dont have best example')
```

```python
with open('sample.txt','w') as fobj:
    print(fobj.readable(),fobj.writable())
    fobj.write('ab')
    print(fobj.read)
```

```python
with open('sample.txt','w+') as fobj:
    print(fobj.readable(),fobj.writable())
    print(fobj.write('12345678'))
    fobj.seek(0)
    print(fobj.read())
```
