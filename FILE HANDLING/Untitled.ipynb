```python
def half(l):
    if len(l) > 1:
        mid = len(l) // 2
        left, right = l[:mid], l[mid:]

        # Recursive sort calls
        half(left)
        half(right)

        # Merge the sorted halves
        merge(l, left, right)


def merge(l, left, right):
    lind, rind, mind = 0, 0, 0

    # Merge until one side is exhausted
    while lind < len(left) and rind < len(right):
        if left[lind] < right[rind]:
            l[mind] = left[lind]
            lind += 1
        else:
            l[mind] = right[rind]
            rind += 1
        mind += 1

    # Copy remaining elements from left
    while lind < len(left):
        l[mind] = left[lind]
        lind += 1
        mind += 1

    # Copy remaining elements from right
    while rind < len(right):
        l[mind] = right[rind]
        rind += 1
        mind += 1


# Input list
l = [11, 16, 23, 43, 0, 1]
half(l)
print(l)
```
