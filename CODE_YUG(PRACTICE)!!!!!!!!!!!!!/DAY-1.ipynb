```python
s=['1234python6648','1234python6648','1234python6648']
dig=[]
char=[]

for item in s:
    int_=''
    dig_=''
    for ele in item:
        if ele.isdigit():
            int_=int_+ele
        else:
            dig_=dig_+ele
    dig.append(int_)
    char.append(dig_)

print(dig)
print(char)
```

```python
num=[78,45,76,23,37,84]
for ind1 in range(len(num)-1,0,-1):
    for ind2 in range(ind1):
        if num[ind2]>num[ind2+1]:
            num[ind2],num[ind2+1]=num[ind2+1],num[ind2]
print(num[-2])
        
```

```python
s='abcaba'
count={}
res=[]
out=''

for ch in s:
    if ch in res:
        count[ch]+=1
    else:
        count[ch]=1
    res.append(ch)
    print(f'{ch}-{count[ch]}')
    out+=ch*count[ch]
print(out)
        
    
```
