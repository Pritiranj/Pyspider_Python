```python
class Holiday(Exception):  # Use uppercase 'H' for class naming conventions
    def __init__(self, spam):  # Correcting the self parameter
        self.spam = spam
    
    def __str__(self):
        return self.spam  # Returning the error message

try:
    raise Holiday('Health problem')  # Raising the custom exception
except Holiday as msg:  # Catching the specific custom exception
    print(msg)  # Printing the exception message
```
