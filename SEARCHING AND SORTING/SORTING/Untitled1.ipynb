```python
#bubble sort

l=[9,31,36,-7,18,0,9]
for ind1 in range(len(l)-1,0,-1):
    for ind2 in range(ind1):
        if l[ind2]>l[ind2+1]:
            l[ind2],l[ind2+1]=l[ind2+1],l[ind2]
print(l)
```

```python
l = [9, 31, 36, -7, 18, 0, 9]

# Bubble sort
for ind1 in range(len(l) - 1, 0, -1):
    for ind2 in range(ind1):
        if l[ind2] > l[ind2 + 1]:  # Corrected condition to compare values
            l[ind2], l[ind2 + 1] = l[ind2 + 1], l[ind2]  # Corrected swapping logic

print(l)
```

```python
#selection sort

l = [9, 31, 36, -7, 18, 0, 9]

for ind1 in range(len(l)-1):
    li=ind1
    for ind2 in range(ind1+1,len(l)):
        if l[ind2]<l[li]:
            li=ind2
    l[ind1],l[li]=l[li],l[ind1]
print(l)
```

```python
l = [9, 31, 36, -7, 18, 0, 9]

# Selection sort
for ind1 in range(len(l) - 1):
    li = ind1  # Assume the smallest element is at index ind1
    for ind2 in range(ind1 + 1, len(l)):  # Compare with the rest of the list
        if l[ind2] < l[li]:  # Find the smallest element
            li = ind2
    # Swap the smallest element with the current element at ind1
    l[ind1], l[li] = l[li], l[ind1]

print(l)
```
