```python
l = [9, 31, 36, -3]
for ind1 in range(len(l)-1):
    li = ind1  # Assume the smallest element is at index ind1
    for ind2 in range(ind1 + 1):
        if l[ind2] < l[li]:  # Find the smallest element in the unsorted part
            li = ind2
        l[ind1], l[li] = l[li], l[ind1]  # Swap the smallest element with the first unsorted element
print(l)
```

```python
def quick(l):
    if len(l)<=1:
        return l
    pivot=l[0]
    left=[ele for ele in (l[1:]) if ele<pivot]
    right=[ele for ele in (l[1:]) if ele>pivot]
    return quick(left)+[pivot]+quick(right)
l=[7,5,11,999,-3,2]
print(quick(l))
```

```python
l=[19,36,24,7]
for ind in range(1,len(l)):
    h=ind
    val=l[h]
    while h>0 and val<l[h-1]:
        l[h]=l[h-1]
        h-=1
    l[h]=val
print(l)
```

```python
l = [19, 36, 24, 7]
for ind in range(1, len(l)):
    h = ind
    val = l[h]
    while h > 0 and val < l[h - 1]:  # Ensure h > 0 to avoid out-of-bounds access
        l[h] = l[h - 1]  # Shift the larger element one position to the right
        h -= 1
    l[h] = val  # Place the current value in its correct position
print(l)
```
