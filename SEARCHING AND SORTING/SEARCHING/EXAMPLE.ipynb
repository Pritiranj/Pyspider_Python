```python
l = [5, 25, 9, 11, 13, 25, 29]
user = 9

def interpolation(l, user, low=0, high=len(l)-1):
    l=sorted(set(l))
    while low <= high and l[low] <= user <= l[high]:  # Fixed l[0] to l[low]
        mid = low + ((high - low) * (user - l[low])) // (l[high] - l[low])  # Fixed formula

        if l[mid] == user:
            return mid
        elif l[mid] < user:
            low = mid + 1
        else:
            high = mid - 1

    return -1  # Removed unnecessary else

print(interpolation(l, user))
```
