```python
#linear search

l=[5,25,9,11,13,25,29]
user=2
for ind in range(len(l)):
    if user==l[ind]:
        print(ind)
        break
else:
    print(-1)
```

```python
#binary search

l=[5,25,9,11,13,25,29]
user=5
res=[]
for ele in l:
    if ele not in res:
        res+=[ele]
res.sort()
low=0
high=len(res)-1
while low<=high :
    mid=(high+low)//2
    if res[mid]>user:
        high=mid-1
    elif res[mid]<user:
        low=mid+1
    elif res[mid]==user:
        print(mid)
        break
else:
    print(-1)
```
