```python
#by using recursion

def linear(l,user,ind=0):
    if ind == len(l):
        return -1
    if user==l[ind]:
        return ind
    return linear(l,user,ind+1)
l=[5,25,34,56]
user=3
print(linear(l,user))
```

```python
l = [5, 11, 2, -4, 7, 6, 24, 9, 9]
res = []
for ele in l:
    if ele not in res:
        res += [ele]
res.sort()

low = 0
high = len(l) - 1
user = 7
while low <= high and res[low] <= user <= res[high]:
    # Calculate the mid index using interpolation formula
    mid = int(low + ((high - low) / (res[high] - res[low]) * (user - res[low])))
    #print('mid=',mid)  # Print the calculated mid index

    # Check the value at mid
    if res[mid] < user:
        low = mid + 1
    elif res[mid]>user:
        high = mid - 1
    elif res[mid] == user:
        print(f"Element found at index {mid}")
        break
else:
    print("Element not found")
```
