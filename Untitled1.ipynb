```python
l=[10,20,5,30,3,4,]
res=[]
for ele in l:
    if ele not in res:
        res+=[ele]
print(res)
res.sort()
print(res)

low=0
high=len(res)-1
user=14

while low <= high:
    mid=(high+low)//2
    if res[mid]<user:
        low=mid+1
    elif res[mid]>user:
        high=mid-1
    elif res[mid]==user:
        print(mid)
        break
else:
    print(-1)
        
    
```
