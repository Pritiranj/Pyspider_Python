```python
from functools import reduce
def arm(num,res=0):
    dup=num
    power=len(str(num))
    while num != 0:
        rem=num%10
        res=res+rem**power
        num//=10
    return res==dup
print(list(reduce(int,arm,range(100,200))))
print([ num for num in range(100,200) if arm(num)])
```

```python
def sqrt(num,res=0):
    while num!=0:
        rem=num%10
        res=res+rem**2
        num//=10
    return res
def happy(num):
    if num>9:
        num=sqrt(num)
    if num==1 or num==7:
        return 'True'
print(list(filter(happy,range(1,200))))
print([num for num in range(1,200) if happy(num)])
```

```python
from functools import reduce
def sample(num1,num2):
    return num1+num2
print(reduce(sample,range(1,5)))
```
