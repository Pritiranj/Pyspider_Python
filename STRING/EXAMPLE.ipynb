```python
#PRINT SUBSTRING IN A STRING


s='madam'

for sv in range(len(s)):
    for ev in range(sv+1,len(s)+1):
        res=''
        for ind in range(sv,ev):
            res+=s[ind]
        print(res)
```

```python
#CHECK SUBSTRING IS PALINDROME OR NOT 

s='madam'

for sv in range(len(s)):
    for ev in range(sv+1,len(s)+1):
        res=''
        for ind in range(sv,ev):
            res+=s[ind]
        rev=''
        for ind in range(len(res)-1,-1,-1):
            rev+=s[ind]
        if res==rev:
            print(res)
```

```python
#REMOVING THE DUPLICATE FROM THE STRING 


s='raining'
res=''

for ch in s:
    if ch not in res:
        res+=ch
print(res)
```

```python
#PRINT OCCURANCES OF EACH CHRACTER


s='raining'
res=''

for ch in s:
    if ch not in res:
        res+=ch
for ch in res:
    print(f'{ch}-{s.count(ch)}')
```

```python
#REVERSE THR POSITION OF A WORD 



s='we are raining in the class'

l=[]
word=''

for ch in s:
    if ch==' ':
        l=[word]+l
        word=''
    else:
        word=ch+word
l=[word]+l
print(l)
```

```python
# CHECK PALINDROME OR NOT WITH LESS NO OF ITREATAION

s='malayalam'

for ind in range(len(s)//2):
    if s[ind] != s[-ind-1]:
        print('not palindrome')
        break
else:
    print('palindrome')
    
```
