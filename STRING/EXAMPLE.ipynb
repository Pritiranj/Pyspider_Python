```python
#REVERSE STRING

s='pushpa2'
res=''
for ch in s:
    res=ch+res
print(res)
```

```python
#PALINDROME 

s='yay'
res=''
for ch in s:
    res=ch+res
if res==s:
    print('palindrome')
else:
    print('not palindrome')
#print(res)
```

```python
#PRINT SUBSTRING IN A STRING


s='madam'

for sv in range(len(s)):
    for ev in range(sv+1,len(s)+1):
        res=''
        for ind in range(sv,ev):
            res+=s[ind]
        print(res)
```

```python
#CHECK SUBSTRING IS PALINDROME OR NOT 

s='madam'

for sv in range(len(s)):
    for ev in range(sv+1,len(s)+1):
        res=''
        for ind in range(sv,ev):
            res+=s[ind]
        rev=''
        for ind in range(len(res)-1,-1,-1):
            rev+=s[ind]
        if res==rev:
            print(res)
```

```python
#REMOVING THE DUPLICATE FROM THE STRING 


s='raining'
res=''

for ch in s:
    if ch not in res:
        res+=ch
print(res)
```

```python
#PRINT OCCURANCES OF EACH CHRACTER


s='raining'
res=''

for ch in s:
    if ch not in res:
        res+=ch
for ch in res:
    print(f'{ch}-{s.count(ch)}')
```

```python
#REVERSE THR POSITION OF A WORD 



s='we are raining in the class'

l=[]
word=''

for ch in s:
    if ch==' ':
        l=[word]+l
        word=''
    else:
        word=ch+word
l=[word]+l
print(l)
```

```python
# CHECK PALINDROME OR NOT WITH LESS NO OF ITREATAION

s='malayalam'

for ind in range(len(s)//2):
    if s[ind] != s[-ind-1]:
        print('not palindrome')
        break
else:
    print('palindrome')
    
```

```python
#AAABBBCCC----->>3A3B3C


s='aaabbbcccd'
res=''
count=1

for ind in range(len(s)-1):
    if  s[ind]==s[ind+1]:
        count+=1
    else:
        res+=str(count)+s[ind]
        count=1
else:
    #res+=str(count)+s[ind+1] 
    res+=str(count)+s[-1]  
print(res)
```

```python
#3a3b3c1d----->AAABBBCCCD


s='3a3b3c1d'
res=''
num=''
for ch in s:
    if '0'<= ch <= '9' :
        num+=ch
    else:
        res+=int(num)*ch
        num=''

print(res)
```
