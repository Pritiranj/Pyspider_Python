```python
num=5
space=0
star=num
for row in range(num):
    for sp in range(space):
        print(' ',end=' ')
    for st in range(star):
        if row==0 or st==0 or st+row==num-1:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
    space+=1
    star-=1
```

```python
s='aaabbbbccd'
l=''
count=1
for ind in range(0,len(s)-1):
    if s[ind]==s[ind+1]:
        count+=1
    else:
        l=l+str(count)+s[ind]
        count=1
else:
    l=l+str(count)+s[ind+1]
print(l)
```
