```python
def arm(num,count):
    if num==0 :
        return 1
    return (num%10)**count+arm(num//10,count-1)
num=135
count=len(str(num))
if (arm(num,count))==num:
    print("arm")
else:
    print(' o')
```

```python
#reverse the number

def rev(num,pos):
    if num==0:
        return 0
    return (num%10)*pos+rev(num//10,pos//10)
num=132
pos=10**(len(str(num))-1)
print(rev(num,pos))
```

```python
#strong number

def fact(num):
    if num==0 :
        return 1
    return num*fact(num-1)
def add(num):
    if num==0:
        return 0
    return fact(num%10)+add(num//10)
def strong(num):
    if num==add(num):
        return 'strong'
    else:
        return 'not strong'
num=145
print(strong(num))
```

```python
def square(num):
    if num==0:
        return 0
    return (num%10)+square(num//10)
def checkdigit(num):
    if num>9:
        return checkdigit(square(num))
    return num
num=7
sidi=checkdigit(num)
if sidi==1 or sidi==7:
    print("yes")
else:
    print("no")
```

```python
def squart(num, res=0):
    while num != 0:
        rem = num % 10
        res = res + rem**2
        num //= 10
    return res

def happy(num):
    if num > 9:  # Keep reducing until num <= 9
        num = squart(num)
    if num == 1 or num == 7:  # Happy numbers end in 1 or 7
        return True
    return False  # Non-happy numbers

print(list(filter(happy, range(1, 1200))))
```
