```python
s='aaabbbbccdde'
res=''
count=1
for ind in range(0,len(s)-1):
    if s[ind]==s[ind+1]:
        count+=1
    else:
        res=res+str(count)+s[ind]
        count=1
else:
    res=res+str(count)+s[ind+1]
print(res)
```

```python
def fact(num,res=1):
    while num != 0:
        res=res*num
        num-=1
    return res
for val in map(fact,range(1,6)):
    print(val)
```
