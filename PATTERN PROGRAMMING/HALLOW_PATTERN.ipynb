```python
num=5
star=num
space=0
for row in range(num):
    for sp in range(space):
        print(' ',end=' ')
    for st in range(star):
        if st==0 or row == 0 or row + st == num-1 :
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
    space+=1
    star-=1
```

```python
num=5
space=num-1
star=1

for row in range(num):
    for sp in range(space):
        print(' ',end=' ')
    for st in range(star):
        if st==0 or row== num-1 or 2*row == st:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
    space-=1
    star+=2
```

```python
num=4
space=0
star=num*2-1

for row in range(num):
    for sp in range(space):
        print(' ',end=' ')
    for st in range(star):
        if row==0 or st==0 or st==(num*2-2)-(row*2):
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
    star-=2
    space+=1
```
