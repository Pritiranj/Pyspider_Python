```python
def con(num,power=1):
    if num==0:
        return 0
    return (num%10)*power + con(num//10,power*2)
num=1110
print(con(num))
```

```python
def sqrt(num,res=0):
    while num != 0:
        rem=num%10
        res=res+rem**2
        num//=10
        return res

def happy(num):
    if num > 9:
        num=sqrt(num)
    if num==1 or num==7:
        return 'happy'
    return 'nh'
# for ele in happy(10,20):
#     print(ele)

for ele in range(10, 20):
    print(f"{ele}")

print(list(filter(happy,range(10,20))))
```

```python
#INT-BINARY(function)


def con(num,res=0,power=1):
    while num != 0:
        rem=num%2
        res=res+rem*power
        power*=10
        num//=2
    return res
num=14
print(con(num))
```

```python
#BINARY-INT(function)

def con(num,power=0,res=0):
    while num != 0:
        rem=num%10
        res= res+rem*(2**power)
        power+=1
        num//=10
    return res
num=1110
print(con(num))
```

```python
#POLYPRIME (function)

def prime(num):
    if num>1:
        for val in range(2,num//2+1):
            if num % val == 0:
                return False
        return True
    return False

def palin(num,dup,res=0):
    while num  != 0:
        rem=num%10
        res=rem+res*10
        num//=10
    return dup==res
def polyprime(num):
    if prime(num) and palin(num,num):
        return 'yes'
    return 'no'

num=12
print(polyprime(num))
    
```

```python
#STRONG NUMBER(function)


def facto(num,fact=1):

    while num !=0:
        fact*=num
        num-=1
    return fact
def adding(num,res=0):
    while num != 0:
        rem=num%10
        res=res+facto(rem)
        num//=10
    return res

def strong(num):
    if adding(num) == num:
        return 'yes'
    return 'no'
num=145
print(strong(num))
```

```python
def palin(num,dup,res=0):
    while num != 0:
        rem= num %10
        res=rem+res*10
        num//=10

    return res
def prime(num):
    if num>1:
        for val in range(2,num//2+1):
            if num % val ==0:
                return False
        return True
    return False

def emrip(num):
    var=palin(num,dup)
    if var != num and prime(num):
        return 'yes'
    return 'no'
num=13
print(emrip(num))
        
```

```python
def palin(num):
    res = 0
    while num != 0:
        rem = num % 10
        res = res * 10 + rem
        num //= 10
    return res

def prime(num):
    if num > 1:
        for val in range(2, num // 2 + 1):
            if num % val == 0:
                return False
        return True
    return False

def emrip(num):
    var = palin(num)
    if var != num and prime(num) and prime(var):
        return 'yes'
    return 'no'

# Testing
for num in [13, 17, 23, 31, 37, 71, 73]:
    print(f"{num} is Emirp? {emrip(num)}")
    print(f"{num}")
```

```python
#ARMSTRONG NUMBER IN A RANGE WITHOUT USING FILTER FUNCTION 


def arm(num,res=0):
    power=len(str(num))
    dup=num
    while num != 0:
        rem=num%10
        res=res+rem**power
        num//=10
    return res==dup
#print(ele for ele in range(100,200) if arm):
print([ele for ele in range(100,150) if arm(ele)])
```
