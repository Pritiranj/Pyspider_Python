```python
import json
```

```python
json.dumps(True)
```

```python
d={'a':[22,34,45],'b':['c',30,29]}
with open('prac.txt','w') as fobj:
    content=json.dumps(d,indent=4)
    fobj.write(content)
```

```python
x={'e':[22,34,45],'f':['c',30,29]}
with open('prac1.txt','w') as fobj:
    json.dump(x,fobj,indent=4)
```

```python
with open('prac1.txt') as fobj:
    content=fobj.read()
    print(type(content))
    pythonobj=json.loads(content)
    print(type(pythonobj))
```

```python
with open('prac.txt') as fobj:
    s=fobj.read
    print(s)
    content=json.load(fobj)
    
    print(content)
    
```
