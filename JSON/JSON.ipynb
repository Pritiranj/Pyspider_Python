PICKLING


```python
import pickle
d={'a':[1,2,3,'hi'],'b':[34,67]}
#d = {'a': [1, 2, 3, 'hi'], 'b': [34,67]}
with open('prac.txt','wb') as fobj:
        con=pickle.dumps(d)
        fobj.write(con)
```

```python
import pickle
d={'a':[1,2,3,4]}
with open('prac.txt','wb') as fobj:
    pickle.dump(d,fobj)
```

```python
import pickle
with open ('prac.txt','rb') as fobj:
    con=fobj.read()
    obj=pickle.loads(con)
    print(obj)
```

```python
import pickle
d={'a':[1,2,3,4]}
with open('prac.txt','wb') as fobj:
    pickle.dump(d,fobj)
```

```python
import pickle
with open ('prac.txt','rb') as fobj:
    data=pickle.load(fobj)
print(data)
```
