```python
#oveloading
class A:
    def m1(self,n1):
        print(n1)
    def m1(self,n1,n2):
        print(n1*n2)
    def m1(self,n1,n2,n3):
        print(n1*n2*n3)
obj=A()
obj.m1(1,2,3)
```

ACHIEVE OVELOADING IN PYTHON

```python
#By using Default Argument
class A:
    def m1(self,num1=0,num2=1,num3=1,num4=1):
        print(num1*num2*num3*num4)
obj=A()
obj.m1(1,2,4)
```

```python
#using variable length not keyword Argument
class A:
    def m1(self,*arg):
        if len(arg)>1:
            res=1
            for ele in arg:
                res*=ele
            print(res)
        else:
            print(0)
obj=A()
obj.m1(1,2,3,4)
```

```python
#over ridding

class A:
    def saturday(self):
        print('no class')
    def sunday(self):
        print('we will come')
class B:
    def saturday(self):
        print('having class')
obj=B()
obj.saturday()
```

```python
class A:
    def saturday(self):
        print('no class')

class B():
    def saturday(self):  # Corrected the method name
        print('having class')

obj = B()
obj.saturday()
```
