```python
#single level inheritance

class parent:
    var1=30
    var2=40
    def m1(self):
        print('m1 of parent')
class child(parent):
    var2=700
    var3=200
    def m2(self):
        print('m1 in child')
obj=child()
obj.m2()
```

```python
  #multilevel inheritance

class A:
    var1=20
class B(A):
    var2=34
class C(B):
    var3=36

obj=C()
print(obj.var1)
```

```python
class A:
    def m1(self):
        print('ya')
class B:
    def m1(self):
        print('hu')
class C(A,B):
    def m1(self):
        B.m1(self)
        super().m1()
        print('c')
obj=C()
obj.m1()
```

```python
class A:
    def __init__(self):
        print('ya')

class B:
    def __init__(self):
        print('hu')

class C(A, B):
    def __init__(self):
        # Call the constructor of class A
        A.__init__(self)
        # Call the constructor of class B
        B.__init__(self)
        # Print from class C
        print('c')

# Create an object of class C
obj = C()
```
