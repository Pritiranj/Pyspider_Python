```python
class sample():
    a=20 
    b=40
    @classmethod
    def m1(cls):
        cls.a=40
obj=sample()
print(obj.m1())
sample.m1() 
print(obj.a)
```

```python
class mock:
    def __init__(self):
        self.a=400  #public(anywhere)
        self._b=500  #protected (access within the package)
        self.__c=600 # private(only within the class)
    def m1(self):
        print(self.a)
        print(self._b)
        print(self.__c)
obj=mock()
#obj.m1()
print(obj.a)
print(obj._b)
print(obj.__c)
```

```python
class mock:
    def __init__(self):
        self.a=400  #public(anywhere)
        self._b=500  #protected (access within the package)
        self.__c=600 # private(only within the class)
    def getter(self):
        return self.__c
    def setter(self):
        self.__c=700
    
obj=mock()
#obj.m1()
#print(obj.getter())
obj.setter()
#print(obj.setter())
print(obj.getter())
```
