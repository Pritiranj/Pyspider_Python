```python
from abc import ABC, abstractmethod

class tv(ABC):
    @abstractmethod
    def channel(self):
        pass

    def power(self):
        print('off/on')

class ch1(tv):
    def channel(self):
        print('self1')

class ch2(tv):
    def channel(self):
        print('self2')

obj = ch1()
obj.channel()
```

```python
def outer(cls):
    res = []

    def inner():
        if not res:  # Check if the list is empty
            obj = cls()  # Create an instance of the class
            res.append(obj)
        return res[0]  # Always return the same instance

    return inner

@outer
class A:
    def sample(self):
        print('hello world')

# Call the singleton instance
obj = A()
obj.sample()
```
