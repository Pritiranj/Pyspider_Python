```python
class v1:
    def __init__(self,mob,name):
        self.mob=mob
        self.name=name
    def details(self):
        print(self.mob)
        print(self.name)
class v2(v1):
    def __intt__ (self,mob,name,gender,home): 
        super().__init__(mob,name)
        self.gender=gender
        self.home=home

    def details(self):
        super().details()
        print(self.gender)
        print(self.home)

obj = v2(78945444, 'situ', 'male', 'pratappur') 
obj.details()
```

```python
class v1:
    def __init__(self, mob, name):
        self.mob = mob
        self.name = name

    def details(self):
        print(self.mob)
        print(self.name)

class v2(v1):
    def __init__(self, mob, name, gender, home):  # Fixed constructor name
        super().__init__(mob, name)  # Call parent class constructor
        self.gender = gender
        self.home = home

    def details(self):
        super().details()  # Optionally print parent class details
        print(self.gender)
        print(self.home)

# Corrected object instantiation
obj = v2(78945444, 'situ', 'male', 'pratappur')  
obj.details()
```

```python
#METHOD OVELOADING

class sunday:
    def m1(self,n1):
        print(n1)
    def m1(self,n1,n2):
        print(n1*n2)
    def m1(self,n1,n2,n3):
        print(n1*n2*n3)

obj=sunday()
obj.m1(1,2)
```

```python
#ACHIEVING OVERLOADING---->1

class sunday:
    
    def m1(self,n1=0,n2=1,n3=1):
        print(n1*n2*n3)

obj=sunday()
obj.m1()
```

```python
#METHOD OVERLOADING ----->2

class sunday:
    def m1 (self,*args):
        if len(args)>0:
            res=1
            for ele in args :
                res+=ele
            print(res)
        else:
            print(0)
obj=sunday()
obj.m1(1,2,3)
            
```

```python
# METHOD OVERRIDDING


class mock :
    
    def saturday(self):
        print('good luck')
        
    def sunday (self):
        print('very good')
        
class re_mock(mock):
    
    def saturday (self):
        super().sunday()
        print('keep it up')

obj=re_mock()
obj.saturday()
```

```python
class mock:
    def saturday(self):
        print('good luck')
    
    def sunday(self):
        print('very good')

class re_mock(mock):
    def saturday(self):  # Corrected the typo
        super().sunday()  # Calls the parent class's 'sunday' method
        print('keep it up')

obj = re_mock()
obj.saturday()
```
