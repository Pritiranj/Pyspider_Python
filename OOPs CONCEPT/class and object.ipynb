```python
#class creation

class bike():
    pass
obj=bike()
print(bike)
print(obj)
```

```python
#creating class

class study:
    course='PFSD'
    location='Banglore'
    
#creating obj

situ=study()
banty=study()

#accessing by using obj ref

print(situ.course,banty.location)

# accessing by class ref

print(study.course,study.location)

#modifying by using obj ref

situ.course='python'
print(situ.course,banty.location)

#accessing by using class ref

study.course='mern'
print(study.course)

```
